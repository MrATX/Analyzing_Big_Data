```python
#Importing necessary packages for analytics
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import export_graphviz
from sklearn.ensemble import RandomForestRegressor

#Importing data file and changing data types to increase efficacy of analysis
vg=pd.read_csv('vgsales0.csv')
vg['Platform']=vg['Platform'].astype('category')
vg['Platform_CAT']=vg['Platform'].cat.codes
vg['Year']=vg['Year'].astype('category')
vg['Year_CAT']=vg['Year'].cat.codes
vg['Genre']=vg['Genre'].astype('category')
vg['Genre_CAT']=vg['Genre'].cat.codes
vg['Publisher']=vg['Publisher'].astype('category')
vg['Publisher_CAT']=vg['Publisher'].cat.codes
vg.head()
```

```python
#NA=5, EU=6, JP=7, Other=8, Global=9
#Segmenting data and establishing variable parameters for specific models
#Creating, fitting, and modifying each set of models
#Printing outcome of models via R^2 values

XNA,yNA=vg.iloc[:,[6,7,8,9,10,11,12,13]],vg.iloc[:,5]
XNA_train,XNA_test,yNA_train,yNA_test=train_test_split(XNA,yNA,train_size=0.5,test_size=0.5,random_state=0)
lin_NA=LinearRegression()
lin_NA.fit(XNA_train,yNA_train)
lasso_NA=Lasso(alpha=0.02)
lasso_NA.fit(XNA_train,yNA_train)
ridge_NA=Ridge(alpha=15)
ridge_NA.fit(XNA_train,yNA_train)
dt_NA=DecisionTreeRegressor(random_state=0,max_leaf_nodes=20)
dt_NA.fit(XNA_train,yNA_train)
rf_NA=RandomForestRegressor(random_state=0,n_estimators=125)
rf_NA.fit(XNA_train,yNA_train)
print('North America Sales')
print('Linear R^2 on Train Data:{:.3f}'.format(lin_NA.score(XNA_train,yNA_train)))
print('Linear R^2 on Test Data:{:.3f}'.format(lin_NA.score(XNA_test,yNA_test)))
print('Lasso R^2 on Train Data:{:.3f}'.format(lasso_NA.score(XNA_train,yNA_train)))
print('Lasso R^2 on Test Data:{:.3f}'.format(lasso_NA.score(XNA_test,yNA_test)))
print('Ridge R^2 on Train Data:{:.3f}'.format(ridge_NA.score(XNA_train,yNA_train)))
print('Ridge R^2 on Test Data:{:.3f}'.format(ridge_NA.score(XNA_test,yNA_test)))
print('DT Accuracy on Train Data:{:.3f}'.format(dt_NA.score(XNA_train,yNA_train)))
print('DT Accuracy on Test Data:{:.3f}'.format(dt_NA.score(XNA_test,yNA_test)))
print('RF Accuracy on Train Data:{:.3f}'.format(rf_NA.score(XNA_train,yNA_train)))
print('RF Accuracy on Test Data:{:.3f}'.format(rf_NA.score(XNA_test,yNA_test)))
print('')
XEU,yEU=vg.iloc[:,[5,7,8,9,10,11,12,13]],vg.iloc[:,6]
XEU_train,XEU_test,yEU_train,yEU_test=train_test_split(XEU,yEU,train_size=0.4,test_size=0.6,random_state=0)
lin_EU=LinearRegression()
lin_EU.fit(XEU_train,yEU_train)
lasso_EU=Lasso(alpha=0.0005)
lasso_EU.fit(XEU_train,yEU_train)
ridge_EU=Ridge(alpha=5)
ridge_EU.fit(XEU_train,yEU_train)
dt_EU=DecisionTreeRegressor(random_state=0,max_leaf_nodes=20)
dt_EU.fit(XEU_train,yEU_train)
rf_EU=RandomForestRegressor(random_state=0,n_estimators=200)
rf_EU.fit(XEU_train,yEU_train)
print('Europe Sales')
print('Linear R^2 on Train Data:{:.3f}'.format(lin_EU.score(XEU_train,yEU_train)))
print('Linear R^2 on Test Data:{:.3f}'.format(lin_EU.score(XEU_test,yEU_test)))
print('Lasso R^2 on Train Data:{:.3f}'.format(lasso_EU.score(XEU_train,yEU_train)))
print('Lasso R^2 on Test Data:{:.3f}'.format(lasso_EU.score(XEU_test,yEU_test)))
print('Ridge R^2 on Train Data:{:.3f}'.format(ridge_EU.score(XEU_train,yEU_train)))
print('Ridge R^2 on Test Data:{:.3f}'.format(ridge_EU.score(XEU_test,yEU_test)))
print('DT Accuracy on Train Data:{:.3f}'.format(dt_EU.score(XEU_train,yEU_train)))
print('DT Accuracy on Test Data:{:.3f}'.format(dt_EU.score(XEU_test,yEU_test)))
print('RF Accuracy on Train Data:{:.3f}'.format(rf_EU.score(XEU_train,yEU_train)))
print('RF Accuracy on Test Data:{:.3f}'.format(rf_EU.score(XEU_test,yEU_test)))
print('')
XJP,yJP=vg.iloc[:,[5,6,8,9,10,11,12,13]],vg.iloc[:,7]
XJP_train,XJP_test,yJP_train,yJP_test=train_test_split(XJP,yJP,train_size=0.4,test_size=0.6,random_state=0)
lin_JP=LinearRegression()
lin_JP.fit(XJP_train,yJP_train)
lasso_JP=Lasso(alpha=0.0005)
lasso_JP.fit(XJP_train,yJP_train)
ridge_JP=Ridge(alpha=5)
ridge_JP.fit(XJP_train,yJP_train)
dt_JP=DecisionTreeRegressor(random_state=0,max_leaf_nodes=20)
dt_JP.fit(XJP_train,yJP_train)
rf_JP=RandomForestRegressor(random_state=0,n_estimators=25)
rf_JP.fit(XJP_train,yJP_train)
print('Japan Sales')
print('Linear R^2 on Train Data:{:.3f}'.format(lin_JP.score(XJP_train,yJP_train)))
print('Linear R^2 on Test Data:{:.3f}'.format(lin_JP.score(XJP_test,yJP_test)))
print('Lasso R^2 on Train Data:{:.3f}'.format(lasso_JP.score(XJP_train,yJP_train)))
print('Lasso R^2 on Test Data:{:.3f}'.format(lasso_JP.score(XJP_test,yJP_test)))
print('Ridge R^2 on Train Data:{:.3f}'.format(ridge_JP.score(XJP_train,yJP_train)))
print('Ridge R^2 on Test Data:{:.3f}'.format(ridge_JP.score(XJP_test,yJP_test)))
print('DT Accuracy on Train Data:{:.3f}'.format(dt_JP.score(XJP_train,yJP_train)))
print('DT Accuracy on Test Data:{:.3f}'.format(dt_JP.score(XJP_test,yJP_test)))
print('RF Accuracy on Train Data:{:.3f}'.format(rf_JP.score(XJP_train,yJP_train)))
print('RF Accuracy on Test Data:{:.3f}'.format(rf_JP.score(XJP_test,yJP_test)))
print('')
XOther,yOther=vg.iloc[:,[5,6,7,9,10,11,12,13]],vg.iloc[:,8]
XOther_train,XOther_test,yOther_train,yOther_test=train_test_split(XOther,yOther,train_size=0.4,test_size=0.6,random_state=0)
lin_Other=LinearRegression()
lin_Other.fit(XOther_train,yOther_train)
lasso_Other=Lasso(alpha=0.0001)
lasso_Other.fit(XOther_train,yOther_train)
ridge_Other=Ridge(alpha=1)
ridge_Other.fit(XOther_train,yOther_train)
dt_Other=DecisionTreeRegressor(random_state=0,max_leaf_nodes=10)
dt_Other.fit(XOther_train,yOther_train)
rf_Other=RandomForestRegressor(random_state=0,n_estimators=50)
rf_Other.fit(XOther_train,yOther_train)
print('Other Sales')
print('Linear R^2 on Train Data:{:.3f}'.format(lin_Other.score(XOther_train,yOther_train)))
print('Linear R^2 on Test Data:{:.3f}'.format(lin_Other.score(XOther_test,yOther_test)))
print('Lasso R^2 on Train Data:{:.3f}'.format(lasso_Other.score(XOther_train,yOther_train)))
print('Lasso R^2 on Test Data:{:.3f}'.format(lasso_Other.score(XOther_test,yOther_test)))
print('Ridge R^2 on Train Data:{:.3f}'.format(ridge_Other.score(XOther_train,yOther_train)))
print('Ridge R^2 on Test Data:{:.3f}'.format(ridge_Other.score(XOther_test,yOther_test)))
print('DT Accuracy on Train Data:{:.3f}'.format(dt_Other.score(XOther_train,yOther_train)))
print('DT Accuracy on Test Data:{:.3f}'.format(dt_Other.score(XOther_test,yOther_test)))
print('RF Accuracy on Train Data:{:.3f}'.format(rf_Other.score(XOther_train,yOther_train)))
print('RF Accuracy on Test Data:{:.3f}'.format(rf_Other.score(XOther_test,yOther_test)))
print('')
XGlobal,yGlobal=vg.iloc[:,[5,6,7,8,10,11,12,13]],vg.iloc[:,9]
XGlobal_train,XGlobal_test,yGlobal_train,yGlobal_test=train_test_split(XGlobal,yGlobal,random_state=0)
lin_Global=LinearRegression()
lin_Global.fit(XGlobal_train,yGlobal_train)
lasso_Global=Lasso(alpha=0.0075)
lasso_Global.fit(XGlobal_train,yGlobal_train)
ridge_Global=Ridge(alpha=5)
ridge_Global.fit(XGlobal_train,yGlobal_train)
dt_Global=DecisionTreeRegressor(random_state=0,max_leaf_nodes=20)
dt_Global.fit(XGlobal_train,yGlobal_train)
rf_Global=RandomForestRegressor(random_state=0,n_estimators=150)
rf_Global.fit(XGlobal_train,yGlobal_train)
print('Global Sales')
print('Linear R^2 on Train Data:{:.3f}'.format(lin_Global.score(XGlobal_train,yGlobal_train)))
print('Linear R^2 on Test Data:{:.3f}'.format(lin_Global.score(XGlobal_test,yGlobal_test)))
print('Lasso R^2 on Train Data:{:.3f}'.format(lasso_Global.score(XGlobal_train,yGlobal_train)))
print('Lasso R^2 on Test Data:{:.3f}'.format(lasso_Global.score(XGlobal_test,yGlobal_test)))
print('Ridge R^2 on Train Data:{:.3f}'.format(ridge_Global.score(XGlobal_train,yGlobal_train)))
print('Ridge R^2 on Test Data:{:.3f}'.format(ridge_Global.score(XGlobal_test,yGlobal_test)))
print('DT Accuracy on Train Data:{:.3f}'.format(dt_Global.score(XGlobal_train,yGlobal_train)))
print('DT Accuracy on Test Data:{:.3f}'.format(dt_Global.score(XGlobal_test,yGlobal_test)))
print('RF Accuracy on Train Data:{:.3f}'.format(rf_Global.score(XGlobal_train,yGlobal_train)))
print('RF Accuracy on Test Data:{:.3f}'.format(rf_Global.score(XGlobal_test,yGlobal_test)))
print('')
XNA0,yNA0=vg.iloc[:,[10,11,12,13]],vg.iloc[:,5]
XNA0_train,XNA0_test,yNA0_train,yNA0_test=train_test_split(XNA0,yNA0,random_state=0)
rf_NA0=RandomForestRegressor(random_state=0,n_estimators=50)
rf_NA0.fit(XNA0_train,yNA0_train)
print('North America Sales - Solo')
print('RF Accuracy on Train Data:{:.3f}'.format(rf_NA0.score(XNA0_train,yNA0_train)))
print('RF Accuracy on Test Data:{:.3f}'.format(rf_NA0.score(XNA0_test,yNA0_test)))
print('')
XEU0,yEU0=vg.iloc[:,[10,11,12,13]],vg.iloc[:,6]
XEU0_train,XEU0_test,yEU0_train,yEU0_test=train_test_split(XEU0,yEU0,random_state=0)
rf_EU0=RandomForestRegressor(random_state=0,n_estimators=50)
rf_EU0.fit(XEU0_train,yEU0_train)
print('Europe Sales - Solo')
print('RF Accuracy on Train Data:{:.3f}'.format(rf_EU0.score(XEU0_train,yEU0_train)))
print('RF Accuracy on Test Data:{:.3f}'.format(rf_EU0.score(XEU0_test,yEU0_test)))
print('')
XJP0,yJP0=vg.iloc[:,[10,11,12,13]],vg.iloc[:,7]
XJP0_train,XJP0_test,yJP0_train,yJP0_test=train_test_split(XJP0,yJP0,random_state=0)
rf_JP0=RandomForestRegressor(random_state=0,n_estimators=50)
rf_JP0.fit(XJP0_train,yJP0_train)
print('Japan Sales - Solo')
print('RF Accuracy on Train Data:{:.3f}'.format(rf_JP0.score(XJP0_train,yJP0_train)))
print('RF Accuracy on Test Data:{:.3f}'.format(rf_JP0.score(XJP0_test,yJP0_test)))
print('')
XOther0,yOther0=vg.iloc[:,[10,11,12,13]],vg.iloc[:,8]
XOther0_train,XOther0_test,yOther0_train,yOther0_test=train_test_split(XOther0,yOther0,random_state=0)
rf_Other0=RandomForestRegressor(random_state=0,n_estimators=50)
rf_Other0.fit(XOther0_train,yOther0_train)
print('Other Sales - Solo')
print('RF Accuracy on Train Data:{:.3f}'.format(rf_Other0.score(XOther0_train,yOther0_train)))
print('RF Accuracy on Test Data:{:.3f}'.format(rf_Other0.score(XOther0_test,yOther0_test)))
print('')
XGlobal0,yGlobal0=vg.iloc[:,[10,11,12,13]],vg.iloc[:,9]
XGlobal0_train,XGlobal0_test,yGlobal0_train,yGlobal0_test=train_test_split(XGlobal0,yGlobal0,random_state=0)
rf_Global0=RandomForestRegressor(random_state=0,n_estimators=50)
rf_Global0.fit(XGlobal0_train,yGlobal0_train)
print('Global Sales - Solo')
print('RF Accuracy on Train Data:{:.3f}'.format(rf_Global0.score(XGlobal0_train,yGlobal0_train)))
print('RF Accuracy on Test Data:{:.3f}'.format(rf_Global0.score(XGlobal0_test,yGlobal0_test)))
```
